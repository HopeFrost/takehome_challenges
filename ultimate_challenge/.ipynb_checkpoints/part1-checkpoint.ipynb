## Load and Inspect Data

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
# import data
logins_df = pd.read_json('data/logins.json')
```

```python
# make sure logins are datetimes
logins_df.info()
```

## Aggregate Data into 15 Minute Intervals

```python
# creat a count column to track samples in each 15 min interval
logins_df['count'] = 1
```

```python
# resample data
logins_df = logins_df.set_index('login_time').resample('15min').count()
```

```python
# preview data to check it
logins_df
```

## Visual Exploration of Data

```python
# create a function to quickly visualize the frequency of logins
def plot_weekends(login_data, label, interval):
    """"plot time series with weekends highlighted"""
    fig =plt.figure(figsize=(12,8))
    ax = sns.lineplot(x=login_data.index, y='count', data= login_data)

    #highlight weekends
    xmin, xmax = ax.get_xlim()
    # range of days in date units
    days = np.arange(np.floor(xmin), np.ceil(xmax)+2) 
    weekends = [(dt.weekday()>=5)|(dt.weekday()==0) for dt in mdates.num2date(days)]
    ax.fill_between(days, *ax.get_ylim(), where=weekends, facecolor='k', alpha=.1)
    # set limits back to default values
    ax.set_xlim(xmin, xmax) 
    
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=interval))
    plt.gcf().autofmt_xdate()

    plt.title( '{} 1970 Logins with Weekends Highlighted'.format(label))
```

```python
# split data into months
jan = logins_df['1970-01-01':'1970-01-31']
feb = logins_df['1970-02-01':'1970-02-28']
march = logins_df['1970-03-01':'1970-03-31']
apr = logins_df['1970-04-01':'1970-04-15']
```

```python
# plot hourly averages
plot_weekends(logins_df.rolling(window=4).mean(), 'Hourly average Longings form Jan-01 Through Apr 15th', 5)

# a little extra formating need on the whole set


```

The spikes tend to be Friday and Saturday nights. 

There may be an up trend.

```python
jan.describe(), plot_weekends(jan, "Jan", 2)
```

```python
feb.describe(), plot_weekends(feb, "Feb", 2)
```

```python
march.describe(), plot_weekends(march, "March", 2)
```

```python
apr.describe(), plot_weekends(apr, "April", 2)
```

The Mean goes up each month peeking in March. There is a daily ebb and surge. It seems to spike in the evenings.

```python
# double check the trend with seasonal_decompose
import statsmodels as sm
from statsmodels.tsa import seasonal
from pylab import rcParams
rcParams['figure.figsize'] = 12, 8
decomposition = sm.tsa.seasonal.seasonal_decompose(logins_df['count'], freq=96)
fig = decomposition.plot()
```

We do indeed see an upward trend that seems to drop the last week of the collected data.

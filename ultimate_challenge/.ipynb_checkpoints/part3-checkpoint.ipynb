### Preprocess 
We'll use a Random Forest Classifier so no need to scale the data.

Other options would be KNN(has trouble with higher dementions) or Logistic regression(low correlation in this set might affect this model)

```python
# split X and y

X = riders_df.drop(['retained', 'last_trip_date', 'signup_date'], axis=1)
y = riders_df.retained

X = pd.get_dummies(X)
```

```python
# import modeling and metric libraries
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from  sklearn.model_selection import GridSearchCV
```

```python
# split train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.30, random_state=68)
```

### Modeling

```python
# tune paramiters
RFC_params = {
    'n_estimators': [200, 400, 600],
    'max_depth': [10, 20, 30, 40],
    'min_samples_split': [2, 5, 10, 15],
    'min_samples_leaf': [1, 2, 5, 10],
}
RF = RandomForestClassifier(random_state=68)
RFGS = GridSearchCV(RF, RFC_params)
RFGS.fit(X_train, y_train)
RFGS.best_params_
```

```python
# create model and make predictions

rf = RandomForestClassifier(max_depth= 20,
                            min_samples_leaf= 10,
                            min_samples_split= 2,
                            n_estimators= 600,
                            random_state=68)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
# check scores
print(classification_report(y_test, y_pred))
```

The model predicts the retained users with %80 percision, and %86 recall

```python
# visualize the recall with an ROC plot
from sklearn import metrics

fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred)

auc = metrics.roc_auc_score(y_test, y_pred)

plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
```

```python
# find feature the model weited as more important

importance = rf.feature_importances_
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
    
plt.bar([x for x in range(len(importance))], importance)
```

The feature the of most importance is the avg_rating_by_driver, followed by surge_pct, weekday_pct, and city_King's Landing

```python
train_pred = rf.predict(X_train)
print(classification_report(y_train, train_pred))
```

The model only predicts slightly better on the train set so its not over fitting.
